```python
# import necessary libraries
import pandas as pd
import os
import glob
  
path = os.getcwd() # This command will get you right into the folder where notebooks is being created.
csv_files = glob.glob(os.path.join(path, "*.csv")) # This is to take all the files.

mdf = [] #Creating an empty list

# To take all the files one by one into one place.
for f in csv_files:
    df = pd.read_csv(f) # read the csv file. f means taking file one by one.
    mdf.append(df) # Appending the file into one place but keeping a separator between the files.
    print('Location:', f) #This is just to see, where the files is being taken from.
    print('File Name:', f.split("\\")[-1]) #Again this is just to see file name.
        
df_master = pd.concat(mdf, axis=0, ignore_index=True) #Finally removing the separator and dumping all the data into one file.
```
